# colormaps

Existing colormaps: 
https://matplotlib.org/stable/users/explain/colors/colormaps.html

```python
from matplotlib import colormaps
from visiontext.vis.colormaps import (
    redgreen_bright,
    redgreen_dark,
    rbg_bright,
    rbg_dark,
    rainbow_bright,
    rainbow_dark,
    create_colormap_for_dark_background,
    get_listed_colormap_from_default_cycle,
)

cmaps = {
    "redgreen_bright": redgreen_bright,
    "redgreen_dark": redgreen_dark,
    "rbg_bright": rbg_bright,
    "rbg_dark": rbg_dark,
    "rainbow_bright": rainbow_bright,
    "rainbow_dark": rainbow_dark,
    "rainbow_custom": create_colormap_for_dark_background(
        n=512, inverted=False, scale=0.5, alpha=0.5, add_brightness=0.3
    ),
    "default": get_listed_colormap_from_default_cycle(),
    "turbo": colormaps["turbo"],
}
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a gradient image based on the colormap
gradient = np.linspace(0, 1, 256)
gradient = np.vstack((gradient, gradient))

for cmap_name, cmap in cmaps.items():
    print()

    # Plotting
    plt.figure(figsize=(6, 1))
    plt.imshow(gradient, aspect="auto", cmap=cmap)
    plt.title(f"Cmap {cmap_name} {type(cmap).__name__}")
    plt.axis("off")
    plt.show()
```
